```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gas_conc = pd.read_csv('filter_pack_concentrations_weekly_cleaned.csv')
```

```python
gas_by_place = gas_conc.groupby(['county', 'state_code'])
for key, item in gas_by_place:
    print(gas_by_place.get_group(key), "\n\n")
```

```python
from datetime import datetime

for key, item in gas_by_place:
    for date in sorted(item['date_on'].values):
        pass
#         print(date)
    print(item[['date_on', 'NHNO3']].values)
```

```python
gases = ['TSO4', 'TNH4', 'CA', 'MG', 'NA', 'K', 'CL', 'NSO4', 'NHNO3', 'WSO2', 'WNO3']
columns = ['date_on'] + gases

place = ('Clallam', 'WA')

data = gas_by_place.get_group(place)
```

```python
data = data[columns]
data
```

```python
from cycler import cycler

colors = []
for i in range(len(gases)):
    colors.append('C'+str(i))
color_cycler = cycler(color='bgrcmyk')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

plt.rc('axes', prop_cycle=color_cycler)

for gas in ['TSO4', 'TNH4', 'NSO4', 'NHNO3', 'WSO2', 'WNO3']:
    plt.plot(data['date_on'].values, data[gas].values, label=gas)

ax.invert_yaxis()  # labels read top-to-bottom
# label bars with counts
plt.xlabel('Time')
plt.ylabel('Concentration (uq/m^3)')
ax.set_title('Gas concentration over time in ' + place[0] + ', ' + place[1])
# plt.tight_layout()

# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.9, box.height])

# Put a legend to the right of the current axis
fig.legend()

plt.show()
fig.savefig('assets/gas_over_time2.png', dpi = 300)
```
